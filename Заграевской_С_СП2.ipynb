```python
#Пункт 4
import numpy as np

def solve_stationary_distribution(P):
    # Проверка, что матрица P является квадратной
    if P.shape[0] != P.shape[1]:
        raise ValueError("Матрица P должна быть квадратной")

    # Уравнение для стационарного распределения: p1 P = p1, p2 P = p2
    # p1 + p2 = 1
    A = np.vstack((P.T - np.eye(len(P)), np.ones(len(P))))#1

    # Вектор правой части
    b = np.zeros(len(P) + 1)
    b[-1] = 1

    # Решение системы линейных уравнений
    solution = np.linalg.lstsq(A, b, rcond=None)[0] #2

    # Нормировка, чтобы получить вероятностное распределение
    stationary_distribution = solution / np.sum(solution)#3

    return stationary_distribution

# a) Матрица P для случая A
P_A = np.array([[0.19, 0.81],
                [0.32, 0.68]])

# Решение для случая A
solution_A = solve_stationary_distribution(P_A)
print("Решение для случая A (p1, p2):", solution_A)

# b) Матрица P для случая B
P_B = np.array([[0.26, 0.11, 0.12, 0.15],
                [0.18, 0.18, 0.18, 0.26],
                [0.11, 0.12, 0.15, 0.18],
                [0.18, 0.18, 0.26, 0.11]])

# Решение для случая B
solution_B = solve_stationary_distribution(P_B)
print("Решение для случая B (p1, p2, p3, p4):", solution_B)

```

```python
#Пункт 5
from numpy.random import uniform
import numpy as np

matrixA = np.array([[0.19, 0.81],
                [0.32, 0.68]])

matrixB = np.array([[0.26, 0.11, 0.12, 0.15],
                [0.18, 0.18, 0.18, 0.26],
                [0.11, 0.12, 0.15, 0.18],
                [0.18, 0.18, 0.26, 0.11]])


def predict(matrix):#смоделированное
    steps = 5000 #кол-во шагов в каждой итерации
    count = 5000 #количество генераций 2-3 минуты

    d = {}  # создаем пустой словарь для подсчета количества посещений каждого состояния
    # Инициализация словаря, все состояния начинаются с 0 посещений
    for state in range(len(matrix)):   #цикл for идет по всем состояниям цепи Маркова-равное количеству строк в матрице).
        d[state] = 0#Для каждого состояния state в цикле устанавливается начальное значение в словаре d. В данном случае, значение устанавливается в 0, что предполагает, что изначально ни одно состояние не было посещено

    for _ in range(count):#Это внешний цикл for, который повторяется count раз- обеспечивает повторяемость симуляций,
        cur = 0#Это устанавливает начальное состояние cur в 0 перед каждой новой генерацией цепи Маркова
        for _ in range(steps):#цикл for, который повторяется steps раз. Этот цикл представляет шаги, которые цепь Маркова выполняет в ходе симуляции.
            line = matrix[cur]# переменной line присваивается значение, которое представляет собой текущую строку матрицы matrix для текущего состояния cur. В контексте цепи Маркова, эта строка матрицы представляет собой вероятности перехода из текущего состояния в другие состояния.
            p = uniform(0, 1)# генерируется случайное число p из равномерного распределения на интервале [0, 1].
            sum = 0#переменная будет использоваться для накопления вероятностей при проходе по строке матрицы.
            for i in range(len(line)):# цикла, который проходит по всем элементам текущей строки матрицы line. len(line) возвращает количество элементов в этой строке.
                sum += line[i]#при проходе по элементу line[i], значение этого элемента прибавляется к текущей сумме sum. Это создает кумулятивные вероятности: sum становится суммой вероятностей всех предыдущих элементов и текущего.
                if p < sum:#Если случайное число p, сгенерированное ранее, меньше чем текущая сумма sum, то выполняется следующий блок кода.
                    cur = i#Если условие p < sum истинно, то переменной cur присваивается значение i, которое представляет индекс элемента в строке матрицы. Это значение i будет соответствовать состоянию, в которое следует перейти
                    break# Это обеспечивает выход из цикла, поскольку уже принято решение о следующем состоянии.
        d[cur] += 1 #увеличивается значение в словаре d для текущего состояния cur. То есть, после совершения шага цепи Маркова и определения нового состояния, соответствующий счетчик в словаре увеличивается на 1.
    for state in range(len(matrix)):#Для каждого состояния в цикле
        print(d[state] / count, end=' ')#вычисляется отношение количества посещений (значение в словаре d) к общему числу генераций (count).Ni- количество раз, которое генерация завершалась в состоянии i, а N - общее количество генераций (count).
    print()#, после каждой генерации цепи Маркова выводится строка с эмпирическим распределением посещений для всех состояний.

def solve(matrix):#теоретическое распределение
    tr = []# будет содержать транспонированную матрицу matrix
    for j in range(len(matrix[0])):#цикл, который проходит по индексам столбцов матрицы matrix
        tr.append([])#Для каждого столбца добавляется пустой список в tr. Это создает список списков, который будет представлять собой транспонированную матрицу.
        for i in range(len(matrix)):#цикл проходит по индексам строк матрицы matrix
            tr[j].append(matrix[i][j])# значение из транспонированной матрицы tr в столбце j (созданном на предыдущем шаге) заполняется элементом из оригинальной матрицы matrix. Это осуществляет транспонирование матрицы.
    A = []#пустой список A, который будет содержать строки матрицы.
    for i in range(len(tr)):#цикл, который проходит по индексам строк транспонированной матрицы tr.
        A.append([])#добавляется пустой список в A
        for j in range(len(tr[0])):#цикл проходит по индексам столбцов транспонированной матрицы tr
            if i == j:#проверяется условие, что индекс строки i равен индексу столбца j. Если это условие истинно, значит, мы находимся на главной диагонали матрицы.
                A[i].append(1 - tr[i][j])#в список A[i] добавляется элемент, равный 1 - tr[i][j]. Это выполняется для всех элементов на главной диагонали матрицы, а остальные элементы останутся равными 0.
            else:
                A[i].append(-tr[i][j])#в список A[i] добавляется элемент, равный -tr[i][j]. Это выполняется для элементов, не находящихся на главной диагонали матрицы.
    A.pop()#удаляет последнюю строку из матрицы A. В предыдущем коде, когда выполнялось добавление в A, была добавлена строка, которая состояла из единиц (последняя строка, добавленная в предыдущем блоке кода). Теперь эта строка удаляется.
    A.append([1 for _ in range(len(tr[0]))])#добавляет новую строку в матрицу A, которая состоит из единиц. Длина этой строки равна количеству столбцов в транспонированной матрице tr.
    b = [0 for _ in range(len(tr) - 1)]#создают вектор b, который используется в решении линейной системы. Вектор начинается с нулей и имеет длину на единицу меньше количества строк в транспонированной матрице tr.
    b.append(1)#добавляется значение 1 в конец вектора.
    ans = np.linalg.solve(A, b)# для решения линейной системы уравнений, где A - матрица коэффициентов, b - вектор свободных членов. Результат сохраняется в переменной ans и выводится на экран.
    print(ans)

#solve(matrixA)#выведено теоретическое распределение для цепи Маркова,
#solve(matrixB)#числовым представлением решения линейной системы уравнений, построенной на основе матрицы переходов matrixA.
predict(matrixA)#моделирует цепь Маркова
predict(matrixB)
```

```python
def simulate_discrete_variable(matrix, initial_state, num_steps):
    current_state = initial_state#Устанавливаем начальное состояние в текущее состояние. Это начальное состояние передается в функцию как аргумент initial_state.
    states = [current_state]#Создаем список states, в который мы будем записывать последовательность состояний в процессе моделирования. Начальное состояние добавляется в этот список.

    for _ in range(num_steps):#Запускаем цикл, который повторяется num_steps раз. Этот цикл представляет собой шаги моделирования в течение определенного количества времени.
        line = matrix[current_state]#Получаем текущую строку матрицы переходов, которая представляет вероятности перехода из текущего состояния в другие состояния.
        p = uniform(0, 1)# Генерируем случайное число p из равномерного распределения на интервале [0, 1]. Это число будет использоваться для принятия решения о следующем состоянии.
        cumulative_prob = 0#Инициализируем переменную cumulative_prob, которая будет использоваться для накопления вероятностей при проходе по строке матрицы.

        max_attempts = 100
        for i in range(len(line)):#цикл, который проходит по всем элементам текущей строки матрицы. Для каждого элемента:
            cumulative_prob += line[i]#Накапливаем вероятность текущего перехода.

            if p < cumulative_prob:#Проверяем, достигли ли мы или превысили ли мы случайное число p накопленную вероятность.
                current_state = i# Если условие выполняется, то мы принимаем решение перейти в состояние i.
                states.append(current_state)#Добавляем текущее состояние в список states, представляя тем самым последовательность состояний в процессе моделирования.
                break
            else:
    # Если мы не перешли в новое состояние после max_attempts попыток,
    # принудительно переходим в другое состояние (можно выбрать случайное).
                current_state = np.random.choice(len(matrix[current_state]))

    return states
num_steps = 100
initial_state = 0

simulated_states_A = simulate_discrete_variable(matrixA, initial_state, num_steps)
print("Simulated States for matrixA:", simulated_states_A)

simulated_states_B = simulate_discrete_variable(matrixB, initial_state, num_steps)
print("Simulated States for matrixB:", simulated_states_B)
```
